## CA High Schools Who Have Met the Benchmark (ERW & Math)


```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read Data
df = pd.read_csv('./data/sat_2019_ca.csv')
```

```python
df.head()
```

```python
# Get size (samples & features) of the dataset
size = df.shape
print(size)
```

```python
df.describe()
```

```python
df.info()
```

```python
# Convert non-numeric values to NaN:
df['PctERWBenchmark12'] = pd.to_numeric(df['PctERWBenchmark12'], errors='coerce')
df['NumERWBenchmark12'] = pd.to_numeric(df['NumERWBenchmark12'], errors='coerce')
df['PctMathBenchmark12'] = pd.to_numeric(df['PctMathBenchmark12'], errors='coerce')
df['TotNumBothBenchmark12'] = pd.to_numeric(df['TotNumBothBenchmark12'], errors='coerce')
df['NumMathBenchmark12'] = pd.to_numeric(df['NumMathBenchmark12'], errors='coerce')

```

```python
#Check 'TotNumBothBenchmark12' df type

print(df['TotNumBothBenchmark12'].dtype)
```

```python
df.info()
```

```python
# Select the columns that you want to measure
schools = df[['SName', 'PctBothBenchmark12']]
```

```python
schools
```

```python
# Get the Highest Percentages
schools_high_bench = df[['SName', 'PctBothBenchmark12']]
schools_high_bench = schools.sort_values('PctBothBenchmark12', ascending=False)
```

```python

schools_high_bench
```

```python
schools_high_bench['PctBothBenchmark12'] = schools_high_bench['PctBothBenchmark12'].astype(str)
```

```python
schools_high_bench
```

```python
# Remove rows containing NaN values
schools_high_bench = schools_high_bench.dropna()
schools_high_bench
```

```python
schools_high_bench['PctBothBenchmark12'] = schools_high_bench['PctBothBenchmark12'].replace('*', np.nan).astype(float)
schools_high_bench
```

```python
# Replace '*' values with NaN
schools_high_bench['PctBothBenchmark12'] = schools_high_bench['PctBothBenchmark12'].replace('*', np.nan)

# Convert PctBothBenchmark12 to float data type
schools_high_bench['PctBothBenchmark12'] = schools_high_bench['PctBothBenchmark12'].astype(float)

# Sort data by PctBothBenchmark12 in descending order
schools_high_bench = schools_high_bench.sort_values('PctBothBenchmark12', ascending=False)

# Select the top 5 schools with the highest PctBothBenchmark12 values
top_5_schools = schools_high_bench.head(5)

# Print the top 5 schools
print(top_5_schools)
```

```python
# Set color scheme for California
ca_color = ['red', 'orange', 'green','blue', 'purple']

# Filter the top 5 high schools
top_schools = schools_high_bench.nlargest(5, 'PctBothBenchmark12')

# Create a bar chart for the top 5 high schools
fig, ax = plt.subplots(figsize=(10, 7))
ax.bar(top_schools['SName'], top_schools['PctBothBenchmark12'], color=ca_color)

# Set chart title and axis labels
ax.set_title('Top 5 CA High Schools with High Benchmark Scores', fontsize=16)
ax.set_xlabel('School Name', fontsize=25, loc= 'center')
ax.set_ylabel('% of Students Meeting Both Benchmarks', fontsize=12)

plt.xticks(rotation=35, ha='right', size=12, va='top')

# Set y-axis limit to 100%
ax.set_ylim([0, 100])

# Show the plot
plt.tight_layout()
plt.savefig('./charts/Top_CA_high')


```
